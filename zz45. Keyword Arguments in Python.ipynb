```python
import pandas as pd
stats = pd.read_csv('/Users/markloessi/Python/Demographic-Data.csv')
stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers','IncomeGroup']
```

```python
#filtering is about rows
```

```python
stats.head()
```

```python
stats.InternetUsers < 2
#basically running a true / false check for each row
```

```python
Filter = stats.InternetUsers < 2 #create an object called Filter
```

```python
#now we get to a 'filter' that we are more commonly used to seeing
stats[Filter]
```

```python
Filter2 = stats.BirthRate > 40
#could do like this stats[Filter2] or
stats[stats.BirthRate > 40]
```

```python
#more filters being combined
#lots of errors with something like stats[stats.BirthRate > 40 and stats.InternetUsers < 2]
Filter = stats.InternetUsers < 2
Filter2 = stats.BirthRate > 40
stats[Filter & Filter2]
#stats[stats.BirthRate > 40 & stats.InternetUsers < 2] this won't work even though looks similar
```

```python
#this will work 
stats[(stats.BirthRate > 40) & (stats.InternetUsers < 2)]
#the () tells Python to do these in two different actions separate from one another
```

```python
#MAC load method number 1
#data from : https://sds-platform-private.s3-us-east-2.amazonaws.com/uploads/P4-Demographic-Data.csv 
import numpy as np
import pandas as pd
stats = pd.read_csv('/Users/markloessi/Python/Demographic-Data.csv')
stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers','IncomeGroup']
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
#Marker size, note there is only a marker = 'shape' element in the lmplot documentation 
##however we can see at the bottom of the documentation additional dictionaries
###{scatter,line}_kws : dictionaries
###    Additional keyword arguments to pass to ``plt.scatter`` and
###    ``plt.plot``.    
### passed as either scatter_kws or line_kws
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False, \
                 hue='IncomeGroup', size=7, scatter_kws={"s":200}  )
#### {} is a dictionary object - s : scalar or array_like, shape (n, ), optional
####The marker size in points**2. Default is rcParams['lines.markersize'] ** 2.
```

```python
#Documentation and explanation 
#Seaborn is built on top of MatPlotLib  
#The scattter_kws actually points you back to MatPlotLib so you can use it's variables.  
```
