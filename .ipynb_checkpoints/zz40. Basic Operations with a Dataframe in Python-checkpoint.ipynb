```python
#Basic Operations with dataframes
```

```python
import pandas as pd
stats = pd.read_csv('/Users/markloessi/Python/Demographic-Data.csv')
stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers','IncomeGroup']
```

```python
#Mathmatical Operations - straight forward
result = stats.BirthRate * stats.InternetUsers
result.head()
```

```python
#add a column and data as a calculation
stats['MyCalc'] = stats.BirthRate * stats.InternetUsers
```

```python
stats.head()
```

```python
#remove a column
stats.head()
```

```python
#drop method
stats.drop('MyCalc', axis=1) #shift tab provides help
```

```python
#lets check someting ;)
stats.head()
```

```python
#note MyCalc is still there
#that's because the drop above doesn't modifiy the dataframe like the add did
#in order to drop from dataframe we'll need to restate the dataframe
stats = stats.drop('MyCalc',1)
```
